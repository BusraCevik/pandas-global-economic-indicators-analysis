```python
# Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load Data

data_path = "../data/raw/dataset.csv"
df=pd.read_csv(data_path)
```

```python
# Quick Look

print("First 5 rows:")
display(df.head())

print("\nData Info:")
df.info()

print("\nSummary Statistics:")
display(df.describe())
```

```python
# Check Missing Values

print("\nMissing Values per Column:")
display(df.isnull())

print("\nMissing Values per Column:")
display(df.isnull().sum())

```

```python
# Column Types & Conversion
# Example: converting 'Year' to datetime

if 'year' in df.columns:
    df['year'] = pd.to_datetime(df['year'], format='%Y')
```

```python
# Basic Plots
# Histogram of GDP



if 'GDP (Current USD)' in df.columns:
    plt.figure(figsize=(10,5))
    sns.histplot(np.log10(df['GDP (Current USD)']), kde=True)
    plt.title("GDP Distribution (log scale)")
    plt.xlabel("log10(GDP)")
    plt.show()
```

```python



# Time series plot for GDP per country

if 'GDP (Current USD)' in df.columns and 'country_name' in df.columns:
    sample_countries = df['country_name'].unique()[:5]
    plt.figure(figsize=(12,6))

    for country in sample_countries:
        subset = df[df['country_name'] == country]

        plt.plot(subset['year'], np.log10(subset['GDP (Current USD)']), label=country)

    plt.title("GDP Trends for Sample Countries (log scale)")
    plt.xlabel("Year")
    plt.ylabel("log10(GDP)")
    plt.legend()
    plt.show()
```
